```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("population.csv")
```

```python
countries_to_plot = [
    'United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France',
    'China', 'India', 'Russia', 'United Kingdom', 'Italy', 'Mexico',
    'South Korea', 'Australia', 'South Africa'
]

filtered_data = df[df['Country Name'].isin(countries_to_plot)]

plt.bar(filtered_data['Country Name'], filtered_data['2021'])
plt.xlabel('Country Name')
plt.ylabel('Population in 2021')
plt.title('Population of selected countries in 2021')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
countries_to_plot = [
    'United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France',
     'Russia', 'United Kingdom', 'Italy', 'Mexico',
    'South Korea', 'Australia', 'South Africa', 'Nigeria', 'Argentina',
    'Egypt', 'Turkey', 'Saudi Arabia', 'Spain', 'Indonesia', 'Pakistan',
    'Bangladesh', 'Vietnam', 'Philippines', 'Thailand', 'Malaysia', 'Ukraine'
]

filtered_data = df[df['Country Name'].isin(countries_to_plot)]

plt.hist(filtered_data['2022'], bins=10, edgecolor='black')
plt.xlabel('Population in 2022')
plt.ylabel('Frequency')
plt.title('Population Distribution of Selected Countries in 2022')
plt.show()
```
